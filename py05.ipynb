```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. 데이터 불러오기
df = pd.read_csv("/kwon/Untitled Folder/iris.csv")


# 2. 특성하고 타겟을 분리
X = df.drop('Species', axis = 1) # 꽃받침, 길이, 넓이
Y = df['Species']
df.head(10)
```

```python
# 3. 훈련 / 테스트 데이터의 분량
x_train, x_test, y_train, y_test = train_test_split(
    X, Y, test_size = 0.2, random_state = 42
)

# 4. 모델학습 (KNN)
from sklearn.neighbors import KNeighborsClassifier # K는 상수

knn = KNeighborsClassifier(n_neighbors = 3) 
knn.fit(x_train, y_train) # fit 적합하게 훈련시켜서 맞춤

# 5. 예측 및 평가
from sklearn.metrics import accuracy_score, classification_report # 정확도 점수
y_pred = knn.predict(x_test)
x_test, y_pred, y_test
print("정확도:", accuracy_score(y_test, y_pred))
print("리포트:", classification_report(y_test, y_pred))

# 6. 새로운 데이터 예측 
import numpy as np 
# feature names 예시 (iris 데이터 기준)
inputs = [5.1, 3.5, 1.4, 0.2]
feature_names = ['Sepal.Length','petal.Length'
                 ,'Species.Length', 'Petal.Width']

new_data = np.array([[5.1, 3.5, 1.4, 0.2]])
pred_class = knn.predict(new_data)
print("예측 클래스:", pred_class[0])


```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report , accuracy_score

# 1. 데이터를 사이킷런 에서 가져오기
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns =iris.feature_names)
df.sample(10)
df['target'] = iris.target
df

# 2. 특성과 타겟 분리
X = df.drop('target', axis=1)
Y = df['target']
X

#3. 학습 / 테스트 데이터 분할
x_train, x_test, y_train, y_test = train_test_split(
    X, Y, test_size = 0.2, random_state = 42
)

# 4. 모델 학습
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)

# 5. 예측 / 평가
y_pred = model.predict(x_test)
print("정확도:", accuracy_score(y_test , y_pred))
print("리포트", classification_report(y_test, y_pred))

# 6. 새로운 데이터 예측
import numpy as np 
# feature names 예시 (iris 데이터 기준)
inputs = [5.1, 3.5, 1.4, 0.2]
feature_names = ['sepal Length(cm)','sepal Width(cm)'
                 ,'petal Length(cm)', 'petal.Length(cm)']


new_data = pd.DataFrame([inputs], columns=feature_names)
pred_model = model.predict(new_data)
print(iris.target_names)
print("예측 클래스:", pred_model[0])
```
